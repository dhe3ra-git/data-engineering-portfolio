```python
# Initializing Spark for
# Sales Data Pipeline
# Author: Dheeraj H Reddy B

from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import broadcast
from pyspark import StorageLevel
import logging

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# Create SparkSession
spark = SparkSession.builder \
    .appName("Sales_Data_Pipeline") \
    .config("spark.sql.adaptive.enabled", "true") \
    .config("spark.sql.shuffle.partitions", "10") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")
print("SparkSession created! ✅")
print(f"Spark version: {spark.version}")
```

```python
# Extract -- Read all CSV files

logger.info("EXTRACT STARTED")

BASE_PATH = "/kaggle/working/bronze"

# Read all tables
sales_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/sales/")

customers_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/customers/")

products_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/products/")

stores_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/stores/")

regions_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/regions/")

promotions_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/promotions/")

returns_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/returns/")

date_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{BASE_PATH}/date_dim/")

# Print row counts
print("="*50)
print("EXTRACT COMPLETE — Row Counts:")
print("="*50)
print(f"Sales transactions : {sales_df.count():,}")
print(f"Customers          : {customers_df.count():,}")
print(f"Products           : {products_df.count():,}")
print(f"Stores             : {stores_df.count():,}")
print(f"Regions            : {regions_df.count():,}")
print(f"Promotions         : {promotions_df.count():,}")
print(f"Returns            : {returns_df.count():,}")
print(f"Date dimension     : {date_df.count():,}")
print("="*50)
print("EXTRACT COMPLETED!")
```

```python
# Data Quality Checks

print("="*50)
print("DATA QUALITY CHECKS")
print("="*50)

def check_quality(df, table_name, critical_cols):
    errors = []
    warnings = []

    # Check nulls
    for col in critical_cols:
        null_count = df.filter(F.col(col).isNull()).count()
        if null_count > 0:
            errors.append(f"{col} has {null_count} nulls!")

    # Check duplicates
    total = df.count()
    distinct = df.dropDuplicates([critical_cols[0]]).count()
    dups = total - distinct
    if dups > 0:
        warnings.append(f"{dups} duplicate rows found!")

    # Check row count
    if total == 0:
        errors.append("Table is EMPTY!")

    # Print results
    print(f"\n{table_name}:")
    print(f"  Total rows    : {total:,}")
    print(f"  Distinct rows : {distinct:,}")

    if errors:
        for e in errors:
            print(f"  ❌ ERROR: {e}")
    elif warnings:
        for w in warnings:
            print(f"  ⚠️  WARNING: {w}")
    else:
        print(f"  ✅ All checks passed!")

    return len(errors) == 0

# Run checks on all tables
check_quality(sales_df, "Sales Transactions",
    ["sale_id", "customer_id", "product_id", "final_amount"])

check_quality(customers_df, "Customers",
    ["customer_id", "first_name", "email"])

check_quality(products_df, "Products",
    ["product_id", "product_name", "unit_price"])

check_quality(stores_df, "Stores",
    ["store_id", "store_name"])

check_quality(regions_df, "Regions",
    ["region_id", "region_name"])

check_quality(returns_df, "Returns",
    ["return_id", "sale_id"])

check_quality(date_df, "Date Dimension",
    ["date_id", "full_date"])

print("\n" + "="*50)
print("DATA QUALITY CHECKS COMPLETE! ✅")
print("="*50)
```

```python
# Transform -- Bronze to Silver layer --

print("="*50)
print("TRANSFORM STARTED")
print("="*50)

# First check actual columns in each table
print("Sales columns:", sales_df.columns)
print("Customers columns:", customers_df.columns)
print("Products columns:", products_df.columns)
print("Stores columns:", stores_df.columns)
print("Regions columns:", regions_df.columns)
print("Promotions columns:", promotions_df.columns)

# ── Clean Sales ──
print("\nCleaning sales_transactions...")
clean_sales = sales_df \
    .dropDuplicates(["sale_id"]) \
    .dropna(subset=["sale_id","customer_id","product_id","final_amount"]) \
    .filter(F.col("final_amount") > 0) \
    .filter(F.col("sale_status") == "Completed") \
    .withColumn("sale_date", F.to_date("sale_date")) \
    .withColumn("year", F.year("sale_date")) \
    .withColumn("month", F.month("sale_date")) \
    .withColumn("day", F.dayofmonth("sale_date")) \
    .withColumn("quarter",
        F.when(F.col("month").between(1,3), "Q1")
         .when(F.col("month").between(4,6), "Q2")
         .when(F.col("month").between(7,9), "Q3")
         .otherwise("Q4")) \
    .withColumn("profit",
        F.col("final_amount") - F.col("tax_amount")) \
    .withColumnRenamed("unit_price", "sale_unit_price")
print(f"Clean sales rows: {clean_sales.count():,} ✅")

# ── Clean Customers ──
print("\nCleaning customers...")
clean_customers = customers_df \
    .dropDuplicates(["customer_id"]) \
    .dropna(subset=["customer_id","email"]) \
    .filter(F.col("is_active") == "Y") \
    .withColumn("full_name",
        F.concat(F.col("first_name"), F.lit(" "), F.col("last_name"))) \
    .withColumnRenamed("city", "customer_city") \
    .withColumnRenamed("state", "customer_state") \
    .select("customer_id","full_name","email","phone",
            "gender","age","customer_city","customer_state",
            "region_id","segment")
print(f"Clean customers rows: {clean_customers.count():,} ✅")

# ── Clean Products ──
print("\nCleaning products...")
clean_products = products_df \
    .dropDuplicates(["product_id"]) \
    .dropna(subset=["product_id","unit_price"]) \
    .filter(F.col("unit_price") > 0) \
    .withColumnRenamed("category", "product_category") \
    .withColumnRenamed("unit_price", "product_unit_price")
print(f"Clean products rows: {clean_products.count():,} ✅")

# ── Clean Stores ──
print("\nCleaning stores...")
clean_stores = stores_df \
    .dropDuplicates(["store_id"]) \
    .dropna(subset=["store_id","store_name"]) \
    .withColumnRenamed("city", "store_city") \
    .withColumnRenamed("state", "store_state") \
    .withColumnRenamed("region_id", "store_region_id")
print(f"Clean stores rows: {clean_stores.count():,} ✅")

# ── Clean Promotions ──
print("\nCleaning promotions...")
clean_promotions = promotions_df \
    .withColumnRenamed("category", "promo_category") \
    .withColumnRenamed("discount_percent", "promo_discount_pct")
print(f"Clean promotions rows: {clean_promotions.count():,} ✅")

# ── Clean Regions ──
print("\nCleaning regions...")
clean_regions = regions_df.dropDuplicates(["region_id"])
print(f"Clean regions rows: {clean_regions.count():,} ✅")

# ── Clean Returns ──
print("\nCleaning returns...")
clean_returns = returns_df \
    .dropDuplicates(["return_id"]) \
    .dropna(subset=["return_id","sale_id"]) \
    .filter(F.col("return_status") == "Approved") \
    .withColumn("return_date", F.to_date("return_date"))
print(f"Clean returns rows: {clean_returns.count():,} ✅")

# ── Join all tables ──
print("\nJoining all tables...")
clean_sales.persist(StorageLevel.MEMORY_AND_DISK)

# Join step by step
step1 = clean_sales.join(
    broadcast(clean_customers), on="customer_id", how="left")

step2 = step1.join(
    broadcast(clean_products), on="product_id", how="left")

step3 = step2.join(
    broadcast(clean_stores), on="store_id", how="left")

step4 = step3.join(
    broadcast(clean_regions), on="region_id", how="left")

enriched_sales = step4.join(
    broadcast(clean_promotions), on="promo_id", how="left")

print(f"Enriched sales rows: {enriched_sales.count():,} ✅")
print(f"Enriched sales columns: {enriched_sales.columns}")
clean_sales.unpersist()

print("\n" + "="*50)
print("TRANSFORM COMPLETE! ✅")
print("="*50)
```

```python
# -- Load -- Write to Silver Layer

print("="*50)
print("LOAD — Writing to Silver Layer")
print("="*50)

SILVER_PATH = "/kaggle/working/silver"

# Write enriched sales — partitioned by year and month!
print("\nWriting enriched sales...")
enriched_sales.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .partitionBy("year", "month") \
    .parquet(f"{SILVER_PATH}/sales/")
print("Enriched sales written! ✅")

# Write dimension tables
print("\nWriting dimension tables...")

clean_customers.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{SILVER_PATH}/customers/")
print("Customers written! ✅")

clean_products.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{SILVER_PATH}/products/")
print("Products written! ✅")

clean_stores.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{SILVER_PATH}/stores/")
print("Stores written! ✅")

clean_returns.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{SILVER_PATH}/returns/")
print("Returns written! ✅")

clean_regions.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{SILVER_PATH}/regions/")
print("Regions written! ✅")

date_df.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{SILVER_PATH}/date_dim/")
print("Date dimension written! ✅")

# Verify silver layer
import os
print("\n" + "="*50)
print("Silver Layer Contents:")
print("="*50)
for folder in os.listdir(SILVER_PATH):
    folder_path = f"{SILVER_PATH}/{folder}"
    if os.path.isdir(folder_path):
        size = sum(
            os.path.getsize(os.path.join(dirpath, f))
            for dirpath, dirnames, filenames
            in os.walk(folder_path)
            for f in filenames
        )
        print(f"{folder:20} → {size/1024:.1f} KB")

print("\n" + "="*50)
print("SILVER LAYER COMPLETE! ✅")
print("="*50)
```

```python
# Silver to Gold Aggregations
print("="*50)
print("SILVER TO GOLD AGGREGATIONS")
print("="*50)

GOLD_PATH = "/kaggle/working/gold"

# Read from Silver
print("\nReading from Silver layer...")
silver_sales = spark.read.parquet(f"{SILVER_PATH}/sales/")
print(f"Silver sales rows: {silver_sales.count():,} ✅")

# ── Gold Table 1: Monthly Sales Summary ──
print("\nCreating monthly sales summary...")
monthly_summary = silver_sales \
    .groupBy("year", "month", "region_name", "zone") \
    .agg(
        F.sum("final_amount").alias("total_sales"),
        F.sum("profit").alias("total_profit"),
        F.sum("discount_amount").alias("total_discount"),
        F.sum("tax_amount").alias("total_tax"),
        F.count("sale_id").alias("total_orders"),
        F.countDistinct("customer_id").alias("unique_customers"),
        F.avg("final_amount").alias("avg_order_value"),
        F.max("final_amount").alias("max_order_value"),
        F.min("final_amount").alias("min_order_value")
    ) \
    .withColumn("total_sales", F.round("total_sales", 2)) \
    .withColumn("total_profit", F.round("total_profit", 2)) \
    .withColumn("avg_order_value", F.round("avg_order_value", 2)) \
    .withColumn("profit_margin_pct",
        F.round(F.col("total_profit") /
                F.col("total_sales") * 100, 2)) \
    .orderBy("year", "month", "region_name")

monthly_summary.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{GOLD_PATH}/monthly_summary/")
print(f"Monthly summary rows: {monthly_summary.count():,} ✅")

# ── Gold Table 2: Product Performance ──
print("\nCreating product performance summary...")
product_summary = silver_sales \
    .groupBy("product_id", "product_name",
             "product_category", "sub_category", "brand") \
    .agg(
        F.sum("final_amount").alias("total_revenue"),
        F.sum("profit").alias("total_profit"),
        F.sum("quantity").alias("total_units_sold"),
        F.count("sale_id").alias("total_orders"),
        F.avg("final_amount").alias("avg_sale_value"),
        F.countDistinct("customer_id").alias("unique_customers")
    ) \
    .withColumn("total_revenue", F.round("total_revenue", 2)) \
    .withColumn("total_profit", F.round("total_profit", 2)) \
    .withColumn("avg_sale_value", F.round("avg_sale_value", 2)) \
    .withColumn("profit_margin_pct",
        F.round(F.col("total_profit") /
                F.col("total_revenue") * 100, 2)) \
    .orderBy(F.col("total_revenue").desc())

product_summary.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{GOLD_PATH}/product_summary/")
print(f"Product summary rows: {product_summary.count():,} ✅")

# ── Gold Table 3: Customer Segment Analysis ──
print("\nCreating customer segment analysis...")
segment_summary = silver_sales \
    .groupBy("segment", "region_name", "year", "month") \
    .agg(
        F.sum("final_amount").alias("total_sales"),
        F.count("sale_id").alias("total_orders"),
        F.countDistinct("customer_id").alias("unique_customers"),
        F.avg("final_amount").alias("avg_order_value"),
        F.sum("discount_amount").alias("total_discount")
    ) \
    .withColumn("total_sales", F.round("total_sales", 2)) \
    .withColumn("avg_order_value", F.round("avg_order_value", 2)) \
    .orderBy("year", "month", "segment")

segment_summary.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{GOLD_PATH}/segment_summary/")
print(f"Segment summary rows: {segment_summary.count():,} ✅")

# ── Gold Table 4: Store Performance ──
print("\nCreating store performance summary...")
store_summary = silver_sales \
    .groupBy("store_id", "store_name",
             "store_type", "store_city", "region_name") \
    .agg(
        F.sum("final_amount").alias("total_sales"),
        F.sum("profit").alias("total_profit"),
        F.count("sale_id").alias("total_orders"),
        F.countDistinct("customer_id").alias("unique_customers"),
        F.avg("final_amount").alias("avg_order_value")
    ) \
    .withColumn("total_sales", F.round("total_sales", 2)) \
    .withColumn("total_profit", F.round("total_profit", 2)) \
    .withColumn("avg_order_value", F.round("avg_order_value", 2)) \
    .orderBy(F.col("total_sales").desc())

store_summary.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{GOLD_PATH}/store_summary/")
print(f"Store summary rows: {store_summary.count():,} ✅")

# ── Gold Table 5: Payment Method Analysis ──
print("\nCreating payment method analysis...")
payment_summary = silver_sales \
    .groupBy("payment_method", "year", "month") \
    .agg(
        F.sum("final_amount").alias("total_sales"),
        F.count("sale_id").alias("total_orders"),
        F.avg("final_amount").alias("avg_order_value")
    ) \
    .withColumn("total_sales", F.round("total_sales", 2)) \
    .withColumn("avg_order_value", F.round("avg_order_value", 2)) \
    .orderBy("year", "month",
             F.col("total_sales").desc())

payment_summary.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet(f"{GOLD_PATH}/payment_summary/")
print(f"Payment summary rows: {payment_summary.count():,} ✅")

print("\n" + "="*50)
print("GOLD LAYER COMPLETE! ✅")
print("="*50)
```

```python
# — Analytics SQL Queries
print("="*50)
print("ANALYTICS SQL QUERIES")
print("="*50)

# Register as temp views
silver_sales.createOrReplaceTempView("sales")
monthly_summary.createOrReplaceTempView("monthly_summary")
product_summary.createOrReplaceTempView("product_summary")
store_summary.createOrReplaceTempView("store_summary")
segment_summary.createOrReplaceTempView("segment_summary")

print("Temp views registered! ✅")
print()

# -- Query 1: Top 5 Regions by Revenue 
print("Q1: Top 5 Regions by Total Revenue")
print("-"*40)
spark.sql("""
    SELECT
        region_name,
        zone,
        ROUND(SUM(total_sales), 2) AS total_revenue,
        SUM(total_orders) AS total_orders,
        ROUND(AVG(avg_order_value), 2) AS avg_order_value,
        ROUND(SUM(total_profit) / SUM(total_sales) * 100, 2) AS profit_margin_pct
    FROM monthly_summary
    GROUP BY region_name, zone
    ORDER BY total_revenue DESC
    LIMIT 5
""").show()

# -- Query 2: Top 10 Best Selling Products 
print("Q2: Top 10 Best Selling Products")
print("-"*40)
spark.sql("""
    SELECT
        product_name,
        product_category,
        brand,
        total_units_sold,
        ROUND(total_revenue, 2) AS total_revenue,
        ROUND(profit_margin_pct, 2) AS profit_margin_pct
    FROM product_summary
    ORDER BY total_revenue DESC
    LIMIT 10
""").show()

# -- Query 3: Monthly Revenue Trend
print("Q3: Monthly Revenue Trend 2024")
print("-"*40)
spark.sql("""
    SELECT
        year,
        month,
        ROUND(SUM(total_sales), 2) AS monthly_revenue,
        SUM(total_orders) AS monthly_orders,
        ROUND(SUM(total_profit), 2) AS monthly_profit
    FROM monthly_summary
    GROUP BY year, month
    ORDER BY year, month
""").show(12)

# -- Query 4: Customer Segment Performance 
print("Q4: Revenue by Customer Segment")
print("-"*40)
spark.sql("""
    SELECT
        segment,
        ROUND(SUM(total_sales), 2) AS total_revenue,
        SUM(total_orders) AS total_orders,
        SUM(unique_customers) AS total_customers,
        ROUND(AVG(avg_order_value), 2) AS avg_order_value
    FROM segment_summary
    GROUP BY segment
    ORDER BY total_revenue DESC
""").show()

# -- Query 5: Top 5 Stores 
print("Q5: Top 5 Performing Stores")
print("-"*40)
spark.sql("""
    SELECT
        store_name,
        store_type,
        store_city,
        region_name,
        total_sales,
        total_orders,
        ROUND(total_profit, 2) AS total_profit
    FROM store_summary
    ORDER BY total_sales DESC
    LIMIT 5
""").show()

# -- Query 6: Payment Method Analysis 
print("Q6: Payment Method Breakdown")
print("-"*40)
spark.sql("""
    SELECT
        payment_method,
        ROUND(SUM(total_sales), 2) AS total_revenue,
        SUM(total_orders) AS total_orders,
        ROUND(AVG(avg_order_value), 2) AS avg_order_value
    FROM segment_summary s
    JOIN sales sal ON s.region_name = sal.region_name
    WHERE sal.payment_method IS NOT NULL
    GROUP BY payment_method
    ORDER BY total_revenue DESC
    LIMIT 6
""").show()

# -- Query 7: Window Function — Running Total 
print("Q7: Monthly Running Total Revenue")
print("-"*40)
spark.sql("""
    WITH monthly AS (
        SELECT
            year, month,
            ROUND(SUM(total_sales), 2) AS monthly_revenue
        FROM monthly_summary
        GROUP BY year, month
    )
    SELECT
        year,
        month,
        monthly_revenue,
        ROUND(SUM(monthly_revenue) OVER (
            ORDER BY year, month
            ROWS BETWEEN UNBOUNDED PRECEDING
            AND CURRENT ROW
        ), 2) AS running_total
    FROM monthly
    ORDER BY year, month
""").show(12)

# -- Query 8: Top Category by Region 
print("Q8: Top Product Category per Region")
print("-"*40)
spark.sql("""
    WITH category_region AS (
        SELECT
            region_name,
            product_category,
            ROUND(SUM(final_amount), 2) AS total_sales,
            RANK() OVER (
                PARTITION BY region_name
                ORDER BY SUM(final_amount) DESC
            ) AS rank
        FROM sales
        GROUP BY region_name, product_category
    )
    SELECT region_name, product_category, total_sales
    FROM category_region
    WHERE rank = 1
    ORDER BY total_sales DESC
""").show()

print("="*50)
print("ALL ANALYTICS QUERIES COMPLETE! ✅")
print("="*50)
```
